```python
#Scaling the Dataset

from sklearn.preprocessing import StandardScaler
import statsmodels.api as sm
x_s=StandardScaler()
y_s=StandardScaler()
```

```python
#Variables for Linear Model
X=bikenorm['registered']
Y=bikenorm['cnt']
X=np.array(X).reshape(-1,1)
Y=np.array(Y).reshape(-1,1)
```

```python
x=x_s.fit_transform(X)
y=y_s.fit_transform(Y)
```

```python
#Model Summary

model1=sm.OLS(Y,X).fit()
predictions=model1.predict(X)

model1.summary()
```

```python
#Libraries for Linear Regression

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
#Linear Model
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size =0.2,random_state=42)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x,y)
```

```python
lr.score(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
y_p=lr.predict(x_test)
```

```python
lr.score(y_p,y_test)
```

```python
error=mean_squared_error(y_test,y_p)
```

```python
from sklearn.linear_model import Lasso
```

```python
Lasso= Lasso(alpha=0.01)
Lasso.fit(x_train,y_train)
```

```python
pred1 = Lasso.predict(x_test)
```

```python
print("Mean square error: %.2f" % mean_squared_error(y_test,pred1))
print("R^2 value: %.2f" % r2_score(y_test, pred1))
```

```python
from sklearn.linear_model import Ridge
```

```python
Ridge= Ridge(alpha=0.01)
Ridge.fit(x_train,y_train)
```

```python
pred2=Ridge.predict(x_test)
```
